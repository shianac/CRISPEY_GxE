```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


plt.rcParams['font.size'] = 7
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.cal'] = 'Arial' 
```

```python
cocl = pd.read_csv('../data_tables/Fitness_files/pool4_fitness_fc.tsv', sep ='\t')
caff = pd.read_csv('../data_tables/Fitness_files/pool7_fitness_fc.tsv',sep = '\t')
fluc = pd.read_csv('../data_tables/Fitness_files/pool9_fitness_fc.tsv',sep='\t')

annos = pd.read_csv('../data_tables/QTL_pool_annotations.tsv',sep='\t')
anno_cocl = annos.merge(cocl, 'inner', 'var_id')
anno_caff = annos.merge(caff,'inner','var_id')
anno_fluc = annos.merge(fluc,'inner','var_id')


anno_caff['abscoef'] = abs(anno_caff['coef'])
anno_cocl['abscoef'] = abs(anno_cocl['coef'])
anno_fluc['abscoef'] = abs(anno_fluc['coef'])
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 7
```

```python

SIZE = 7

plt.rc('axes', titlesize=SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SIZE)    # legend fontsize
plt.rc('figure', titlesize=SIZE)  # fontsize of the figure title
```

```python
fig,axes = plt.subplots(2,3, figsize = (6.4,3.5))
fig.tight_layout()
ax = axes[0,2]
ax2 = axes[1,2]
ax3 = axes[0,1]
ax4 = axes[1,1]
ax5 = axes[0,0]
ax6 = axes[1,0]


tor1 = anno_caff[(anno_caff['Gene']=='TOR1')]
tor1[(tor1['condition']=='CAFF')].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax, color = 'lightblue')
tor1[(tor1['condition']=='P7SC')].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax, color = 'gold')
tor1[(tor1['condition']=='CAFF')&(tor1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax, color = 'blue')
try:
    tor1[(tor1['condition']=='P7SC')&(tor1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax, color = 'orangered')
except:
    pass
# set new limits

low, high = ax.get_ylim()
bound = max(abs(low), abs(high))*1.25
ax.set_ylim(-bound, bound)
ysize = bound*2
xsize = ax.get_xlim()[1]-ax.get_xlim()[0]
ax.arrow(559416,-bound+.03*ysize,7413 ,0,  width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax.annotate('TOR1', xy = (559416+7413/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')  
ax.xaxis.set_label_text('')
ax.yaxis.set_label_text('')
ax.locator_params(tight = True,axis="x", nbins=3)
ax.locator_params(tight = True,axis="y", nbins=5)
ax.set_title('Chr11 TOR1 CAFF QTL')


tor2 = anno_caff[(anno_caff['Gene']=='TOR2')]
tor2[tor2['condition']=='CAFF'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax2, color = 'lightblue')
tor2[(tor2['condition']=='P7SC')].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax2, color = 'gold')
tor2[(tor2['padj']<.01)&(tor2['condition']=='CAFF')].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax2, color = 'blue')

try:
    tor2[(tor2['adjP']<.01)&(tor2['condition']=='P7SC')].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax2, color = 'orangered')
except:
    pass
# set new limits

low, high = ax2.get_ylim()
bound = max(abs(low), abs(high))*1.25
ax2.set_ylim(-bound, bound)
ysize = bound*2
xsize = ax2.get_xlim()[1]-ax2.get_xlim()[0]
ax2.arrow(63359,-bound+.03*ysize,-7425 ,0,  width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax2.annotate('TOR2', xy = (63359-7425/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')   
ax2.xaxis.set_label_text('')
ax2.yaxis.set_label_text('')
ax2.locator_params(tight = True,axis="x", nbins=3)
ax2.locator_params(tight = True,axis="y", nbins=5)
ax2.set_title('Chr11 TOR2 CAFF QTL')




mam3  = anno_cocl[(anno_cocl['Gene']=='MAM3')]
mam3[mam3['condition']=='COCL'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax3, color = 'lightblue')
mam3[mam3['condition']=='P4SC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax3, color = 'gold')
mam3[(mam3['condition']=='COCL')&(mam3['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax3, color = 'blue')
try:
    mam3[(mam3['condition']=='P4SC')&(mam3['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax3, color = 'orangered')
except:
    pass

# set new limits

low, high = ax3.get_ylim()
bound = max(abs(low), abs(high))
ax3.set_ylim(-bound, bound)    
ax3.xaxis.set_label_text('')
ax3.yaxis.set_label_text('')
ysize = bound*2    
xsize = ax3.get_xlim()[1]-ax3.get_xlim()[0]
ax3.arrow(216137,-bound+.03*ysize,-2121 ,0,  width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax3.annotate('MAM3', xy = (216137-2121/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')   
ax3.locator_params(tight = True,axis="x", nbins=3)
ax3.locator_params(tight = True,axis="y", nbins=5)
ax3.set_title(r'Chr15 MAM3 CoCl$_2$ QTL')


cot1 = anno_cocl[(anno_cocl['Gene']=='COT1')]
cot1[cot1['condition']=='COCL'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax4, color = 'lightblue')
cot1[cot1['condition']=='P4SC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax4, color = 'gold')
cot1[(cot1['condition']=='COCL')&(cot1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax4, color = 'blue')
try:
    cot1[(cot1['condition']=='P4SC')&(cot1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax4, color = 'orangered')
except:
    pass

low, high = ax4.get_ylim()
bound = max(abs(low), abs(high))*1.25
ax4.set_ylim(-bound, bound)    
ysize = bound*2  
xsize = ax4.get_xlim()[1]-ax4.get_xlim()[0]
ax4.arrow(907555,-bound+.03*ysize,-1320,0,  width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax4.annotate('COT1', xy = (907555-1320/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')   
ax4.xaxis.set_label_text('')
ax4.yaxis.set_label_text('')
ax4.locator_params(tight = True,axis="x", nbins=3)
ax4.locator_params(tight = True,axis="y", nbins=5)
ax4.set_title(r'Chr15 COT1 CoCl$_2$ QTL')
#every_nth = 2
#for n, label in enumerate(ax4.xaxis.get_ticklabels()):
#    if n % every_nth == 0:
#        label.set_visible(False)


pdr3 = anno_fluc[(anno_fluc['chrom']=='II')&(anno_fluc['Gene']=='PDR3')]
pdr3[pdr3['condition']=='FLUC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax5, color = 'lightblue')
pdr3[pdr3['condition']=='P9SC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax5, color = 'gold')
pdr3[(pdr3['condition']=='FLUC')&(pdr3['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax5, color = 'blue')
try:
    pdr3[(pdr3['condition']=='P9SC')&(pdr3['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax5, color = 'orangered')
except: 
    pass

low, high = ax5.get_ylim()
bound = max(abs(low), abs(high))*1.25
ax5.set_ylim(-bound, bound)    
ysize = bound*2    
xsize = ax5.get_xlim()[1]-ax5.get_xlim()[0]
ax5.arrow(217470,-bound+.03*ysize,2931,0,  width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax5.annotate('PDR3', xy = (217470+2931/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')
ax5.xaxis.set_label_text('')
ax5.yaxis.set_label_text('')
ax5.locator_params(tight = True,axis="x", nbins=3)
ax5.locator_params(tight = True,axis="y", nbins=5)
ax5.set_title('Chr2 PDR3 FLC QTL')


sfl1 = anno_fluc[(anno_fluc['Gene'] =='SFL1')]
sfl1[sfl1['condition']=='FLUC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax6, color = 'lightblue')
sfl1[sfl1['condition']=='P9SC'].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax6, color = 'gold')
sfl1[(sfl1['condition']=='FLUC')&(sfl1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax6, color = 'blue')
try:
    sfl1[(sfl1['condition']=='P9SC')&(sfl1['padj']<.01)].plot.scatter('SNP_chr_pos','coef', s = 3, ax= ax6, color = 'orangered')
except:
    pass

xsize = ax6.get_xlim()[1]-ax6.get_xlim()[0]

ax6.xaxis.set_label_text('')
ax6.yaxis.set_label_text('')

low, high = ax6.get_ylim()
bound = max(abs(low), abs(high))
ax6.set_ylim(-bound, bound)  
ysize = bound*2
ax6.arrow(586981,-bound+.03*ysize,2301,0, width = ysize/12,head_width = ysize/12, head_length = xsize/20,length_includes_head=True)  
ax6.annotate('SFL1', xy = (586981+2301/2,-bound+.01*ysize), color = 'white', fontstyle = 'oblique', weight = 'bold', ha = 'center')
ax6.locator_params(tight = True,axis="x", nbins=3)
ax6.locator_params(tight = True,axis="y", nbins=5)
ax6.set_title('Chr15 SFL1 FLC QTL')



plt.subplots_adjust(hspace = .5)
fig.add_subplot(111, frameon=False)
# hide tick and tick label of the big axis

plt.xlabel('Position in Chromosome', labelpad = 25)
plt.ylabel("Variant Fitness", labelpad = 25)
lgnd = ax3.legend(['Stress FDR\u2265.01', 'SC FDR\u2265.01','Stress FDR<.01', 'SC Media FDR<.01'],loc='lower left', bbox_to_anchor = (2.22,-.51))
lgnd.legendHandles[0]._sizes = [20]
lgnd.legendHandles[1]._sizes = [20]
lgnd.legendHandles[2]._sizes = [20]
lgnd.legendHandles[3]._sizes = [20]


plt.tick_params(labelcolor='none', which='both', top=False, bottom=False, left=False, right=False,labelbottom=False,labelleft=False)
plt.savefig('../GxE_Figures/Figure_2/figure2e_exampleQTLs_FC.svg')
```
