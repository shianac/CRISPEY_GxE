```python
import pandas as pd
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['font.size'] = 10
plt.rcParams['font.family'] = 'Arial'


annotations = pd.read_csv('../data_tables/erg_annos.tsv',sep = '\t', index_col='var_id')
gxe = pd.read_csv('../data_tables/ergosterol_pathway_gxe_interactions.tsv', sep = '\t')
var_fits = pd.read_csv('../data_tables/Fitness_files/ergosterol_pathway_fitness.tsv', sep = '\t')
annotated_gxe = gxe.merge(annotations,'inner','var_id')
annotated_fits = var_fits.merge(annotations,'inner','var_id')
OLIGO_TABLE_PATH = '/home/shian/CRISPEY3_fraserserver/tables/oligos_nonuniq_crispey3_GG_9bp_OLIGO_with_seq_primers.txt'
oli_info = pd.read_csv(OLIGO_TABLE_PATH,'\t')
```

```python
def sign_change_checker(arr):
    a = np.array(arr)
    if len(a[a>0])*len(a[a<0]):
        return True
    return False
```

```python


good_vars =[] 

for i, row in annotated_gxe[(annotated_gxe['padj']<0.01)].iterrows():
    var = row['var_id']
    conds = row['gxe'].split('-')

    if annotated_fits[(annotated_fits['var_id']==var)&(annotated_fits['condition'].isin(conds))]['coef'].prod()<0:
        good_vars.append(var)


annotated_fits['num_conds'] = annotated_fits.groupby('var_id')['coef'].transform('size')

annotated_fits['abscoef'] = abs(annotated_fits['coef'])

annotated_fits['biggestEffect'] = annotated_fits.groupby('var_id')['abscoef'].transform('max')

hit_in_one_cond = annotated_fits[(annotated_fits['num_conds']==6)&(annotated_fits['padj']<0.01)]['var_id'].unique()
annotated_fits['signchange'] = annotated_fits.groupby('var_id')['coef'].transform(sign_change_checker)


not_gxe =set(hit_in_one_cond)-set(good_vars)
print(len(set(hit_in_one_cond)))
print(len(set(good_vars)))
print('Frac GxE:' +str(1-len(not_gxe)/len(hit_in_one_cond)))
```

```python
z = []
    
cuts = [1,.5,.25,.2,.15,.1,.05,0.025,.01,.005,.001,.0005,.0001,.00005,.00001,.000005,.000001]
for i in cuts: 
    good_vars =[] 
    for k, row in annotated_gxe[(annotated_gxe['padj']<i)].iterrows():
        var = row['var_id']
        conds = row['gxe'].split('-')
        if annotated_fits[(annotated_fits['var_id']==var)&(annotated_fits['condition'].isin(conds))]['coef'].prod()<0:
            good_vars.append(var)
            
    fit_fracs = []
    for j in cuts:
        

        hit_in_one_cond = annotated_fits[(annotated_fits['num_conds']==6)&(annotated_fits['padj']<j)]['var_id'].unique()
        not_gxe =set(hit_in_one_cond)-set(good_vars)
        fit_fracs.append(1-len(not_gxe)/len(hit_in_one_cond))
        print('hmm '+str(j))
    z.append(fit_fracs)
    print('done with 1')
```

```python
fig, ax = plt.subplots()
def fmt(x):
    s = f"{x:.3f}"
    if s.endswith("0"):
        s = f"{x:.0f}"
    return s
ax.set_xscale('log')
ax.set_yscale('log')
cp = plt.contourf(x, y, z,levels = np.arange(0,1.001,0.025), vmin = 0, vmax = 1)
plt.colorbar(cp,label = 'Fraction of GxE variants')
d = plt.contour(x,y,z, levels= [0.9365853658536585], colors = ['black'], linewidths = 2)
plt.clabel(d, d.levels,inline=True, fmt =fmt)
plt.clim(0,1)

plt.ylabel('GxE FDR cutoff')
plt.xlabel('Fitness effect FDR cutoff')
plt.savefig('../GxE_Figures/Supplementary_Figures/fig_S5.png', dpi = 600)
```
