## Looking at Genomic Loci with multiple edits

```python
import pandas as pd
```

```python
annotations = pd.read_csv('../data_tables/erg_annos.tsv',sep = '\t')
fits = pd.read_csv('../data_tables/ergosterol_pathway_fitness.tsv',sep='\t')
annotated_fits =  annotations.merge(fits, 'inner', 'var_id')
OLIGO_TABLE_PATH = '/home/shian/CRISPEY3_fraserserver/tables/oligos_nonuniq_crispey3_GG_9bp_OLIGO_with_seq_primers.txt'
oli_info = pd.read_csv(OLIGO_TABLE_PATH,'\t')
```

```python
#Getting annotations for each variant in the ergosterol pool that was measured
all_variants = annotated_fits['var_id'].unique()
all_variant_annotations = annotations[(annotations['var_id'].isin(all_variants))]
```

```python
#getting the number of variants/edits per genomic location
all_variant_annotations['edits_per_location'] = all_variant_annotations.groupby(['chrom','SNP_chr_pos'])['var_id'].transform('size')

#getting genomic locations with multiple edits
multi_edits = all_variant_annotations[all_variant_annotations['edits_per_location']>1]['var_id'].unique()
```
